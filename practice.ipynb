## Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
    ])
```

```python
print(type(sales_data))
print(sales_data.shape)
print(sales_data.size)
```

```python
print(f"Sample data for 1st 3 resturant:\n {sales_data[:3]}")
```

```python
# Slicing array
# ':' = all     'm:n' = slice data from m till n
print(f"Sample data for 1st 3 resturant:\n {sales_data[:, 1:]}")
# In this case: select all from row and for column starts from 1
```

## Total sales per Year

```python
print(f"Total sales from column side:\n {np.sum(sales_data[:, 1:],axis=0)}")
print(f"\nTotal sales from row side:\n {np.sum(sales_data[:, 1:],axis=1)}")
```

## Minimum sales per resturant

```python
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

## Max sales per year

```python
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)
```

### Average sales per resturant

```python
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

### Cumulative sales of all years of resturant

```python
cusum = np.cumsum(sales_data[:, 1:], axis=1)
print(cusum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cusum, axis=0))
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
